```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
from scipy.stats import norm  # Normal distribution

# Input widgets for mean and standard deviation
mean_input = widgets.FloatText(value=0, description='Mean:', style={'description_width': 'initial'})
std_input = widgets.FloatText(value=1, description='Std Dev:', style={'description_width': 'initial'})

# Button to update the plot
update_button = widgets.Button(description='Update Curve', button_style='success')

# Output widget for displaying the plot
plot_output = widgets.Output()

# Function to update the plot
def update_curve(change):
    mean = mean_input.value
    std_dev = std_input.value

    with plot_output:
        plot_output.clear_output(wait=True)

        # Create x values with a constant scale
        x = np.linspace(-100, 100, 1000)
        y = norm.pdf(x, loc=mean, scale=std_dev)

        # Plotting
        plt.figure(figsize=(8, 5))
        plt.plot(x, y, label='Normal Distribution')
        plt.fill_between(x, y, alpha=0.2)
        plt.axvline(mean, color='red', linestyle='--', label='Mean')

        # Set x-axis limits to keep them constant
        plt.xlim(-100, 100)

        # Labels and legends
        plt.title('Visualizing the Normal Bell Curve with Fixed X-axis Scale')
        plt.xlabel('Values')
        plt.ylabel('Probability Density')
        plt.legend()
        plt.grid()
        plt.show()

# Link button click to the update function
update_button.on_click(update_curve)

# Layout for widgets and output
controls = widgets.VBox([mean_input, std_input, update_button])
display(widgets.HBox([controls, plot_output]))

# Display the initial plot
update_curve(None)
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parameters (set default values for demonstration)
average_demand_per_day = 10  # Average daily demand
lead_time = 5  # Lead time in days
demand_std_dev = 2  # Standard deviation of daily demand
holding_cost_per_unit = 0.5  # Holding cost per unit
ordering_cost = 50  # Fixed ordering cost
desired_service_level = 0.95  # Target cycle service level ()



# Function to calculate safety stock
def calculate_safety_stock(service_level, demand_std_dev, lead_time):
    """
    Calculate safety stock based on the desired service level.

    Formula:
    Safety Stock = z * _LT
    where:
    z = z-score corresponding to the desired service level
    _LT = demand_std_dev * sqrt(lead_time)
    """
    z = norm.ppf(service_level)  # Convert service level to z-score
    return z * demand_std_dev * np.sqrt(lead_time)

# Function to calculate reorder point
def calculate_reorder_point(average_demand, lead_time, safety_stock):
    """
    Calculate the reorder point (ROP).

    Formula:
    ROP = (average_demand * lead_time) + safety_stock
    """
    demand_during_lead_time = average_demand * lead_time  # Expected demand during lead time
    return demand_during_lead_time + safety_stock

# Function to calculate reorder quantity (Economic Order Quantity)
def calculate_reorder_quantity(average_demand, ordering_cost, holding_cost):
    """
    Calculate the Economic Order Quantity (EOQ).

    Formula:
    EOQ = sqrt((2 * annual_demand * ordering_cost) / holding_cost)
    """
    annual_demand = average_demand * 365  # Convert daily demand to annual demand
    return np.sqrt((2 * annual_demand * ordering_cost) / holding_cost)

# Function to calculate cycle service level ()
def calculate_cycle_service_level(safety_stock, demand_std_dev, lead_time):
    """
    Calculate the Cycle Service Level (probability of no stockout during a cycle).

    Formula:
     = (z), where z = Safety Stock / _LT
    """
    z = safety_stock / (demand_std_dev * np.sqrt(lead_time))  # z-score
    return norm.cdf(z)  # Convert z-score to probability

# Function to calculate fill rate ()
def calculate_fill_rate(safety_stock, demand_std_dev, lead_time, average_demand):
    """
    Calculate Fill Rate (): the proportion of demand fulfilled without backorders.

    Formula:
    Fill Rate = 1 - (expected shortage / total demand)
    """
    sigma_lt = demand_std_dev * np.sqrt(lead_time)  # Standard deviation of demand during lead time
    z = safety_stock / sigma_lt  # z-score
    shortage = sigma_lt * (norm.pdf(z) - z * (1 - norm.cdf(z)))  # Expected shortage
    expected_demand = average_demand * lead_time  # Total demand during lead time
    return 1 - (shortage / expected_demand)

# Function to plot safety stock vs service level
def plot_service_level_vs_safety_stock(demand_std_dev, lead_time):
    """
    Plot the relationship between service level and safety stock.
    """
    service_levels = np.linspace(0.8, 0.99, 20)  # Range of service levels
    safety_stocks = [calculate_safety_stock(sl, demand_std_dev, lead_time) for sl in service_levels]
    plt.figure(figsize=(10, 6))
    plt.plot(service_levels, safety_stocks, marker='o')
    plt.title("Safety Stock vs Desired Service Level")
    plt.xlabel("Desired Service Level ()")
    plt.ylabel("Safety Stock (units)")
    plt.grid()
    plt.show()

# Function to plot fill rate vs safety stock
def plot_fill_rate_vs_safety_stock(demand_std_dev, lead_time, average_demand):
    """
    Plot the relationship between fill rate and safety stock.
    """
    safety_stock_values = np.linspace(0, 50, 20)  # Range of safety stock values
    fill_rates = [calculate_fill_rate(ss, demand_std_dev, lead_time, average_demand) for ss in safety_stock_values]
    plt.figure(figsize=(10, 6))
    plt.plot(safety_stock_values, fill_rates, marker='o')
    plt.title("Fill Rate vs Safety Stock")
    plt.xlabel("Safety Stock (units)")
    plt.ylabel("Fill Rate ()")
    plt.grid()
    plt.show()

# Validate inputs


# Perform calculations
safety_stock = calculate_safety_stock(desired_service_level, demand_std_dev, lead_time)
reorder_point = calculate_reorder_point(average_demand_per_day, lead_time, safety_stock)
reorder_quantity = calculate_reorder_quantity(average_demand_per_day, ordering_cost, holding_cost_per_unit)
cycle_service_level = calculate_cycle_service_level(safety_stock, demand_std_dev, lead_time)
fill_rate = calculate_fill_rate(safety_stock, demand_std_dev, lead_time, average_demand_per_day)

# Display results
print("Inventory Management Metrics:")
print(f"Safety Stock: {safety_stock:.2f} units")
print(f"Reorder Point: {reorder_point:.2f} units")
print(f"Reorder Quantity (EOQ): {reorder_quantity:.2f} units")
print(f"Cycle Service Level (): {cycle_service_level:.2%}")
print(f"Fill Rate (): {fill_rate:.2%}")

# Generate visualizations
plot_service_level_vs_safety_stock(demand_std_dev, lead_time)
plot_fill_rate_vs_safety_stock(demand_std_dev, lead_time, average_demand_per_day)
```

```python
# @title
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from scipy.stats import norm

# -------------------------------
# Step 1: Generate Raw Transactional Data
# -------------------------------

# Set random seed for reproducibility
np.random.seed(42)

num_skus = 50
num_orders_per_sku = 5  # Each SKU will have 5 purchase order records

# Generate SKU IDs
sku_ids = [f"SKU-{i+1:03d}" for i in range(num_skus)]

# We will simulate orders over a given timeframe, say a 6-month period
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 6, 30)

# Generate a range of possible order dates
date_range = pd.date_range(start_date, end_date, freq='D')
date_range_length = len(date_range)

records = []
for sku in sku_ids:
    # For each SKU, create several orders
    for _ in range(num_orders_per_sku):
        order_date = np.random.choice(date_range)

        # Quantity ordered between 50 and 500 units
        quantity_ordered = np.random.randint(50, 501)

        # Lead time between 2 and 10 days
        lead_time_days = np.random.randint(2, 11)

        # Received date is order_date + lead_time_days
        received_date = order_date + pd.Timedelta(days=lead_time_days)

        records.append({
            "SKU_ID": sku,
            "Order_Date": order_date,
            "Quantity_Ordered": quantity_ordered,
            "Received_Date": received_date,
            "Lead_Time_Days": lead_time_days
        })

df = pd.DataFrame(records)

# Sort by SKU and Order_Date for neatness
df.sort_values(by=["SKU_ID", "Order_Date"], inplace=True)
df.reset_index(drop=True, inplace=True)

# -------------------------------
# Step 2: Explaining the Columns
# -------------------------------
# SKU_ID: Unique identifier for the product.
# Order_Date: Date on which the purchase order was placed.
# Quantity_Ordered: How many units were ordered on that date.
# Received_Date: The date when the ordered quantity arrived.
# Lead_Time_Days: The difference in days between when the order was placed and when it was received.

print("my first dataset")


# -------------------------------
#
df
```

```python

import pandas as pd
import numpy as np
from scipy.stats import norm

# Step 1: Group data by SKU
# Theory: Grouping by SKU allows us to calculate metrics like average demand and variability (std dev) for each SKU separately.
sku_groups = df.groupby("SKU_ID")

# Step 2: Calculate Average Daily Demand per SKU
# Theory: Average daily demand helps estimate the typical daily consumption of an SKU, crucial for inventory planning.
avg_daily_demand = sku_groups["Quantity_Ordered"].sum() / (end_date - start_date).days
# Rename for clarity in the final metrics DataFrame
avg_daily_demand = avg_daily_demand.rename("Average_Daily_Demand")

# Step 3: Calculate Demand Standard Deviation per SKU
# Theory: Standard deviation measures the variability in demand, which helps determine safety stock to account for uncertainty.
demand_std = sku_groups["Quantity_Ordered"].std().fillna(0)  # If there's only one order, std is set to 0
demand_std = demand_std.rename("Demand_Std_Dev")

# Step 4: Merge Average Daily Demand and Demand Standard Deviation
# Theory: Combining these metrics into a single DataFrame allows us to calculate further inventory-related metrics.
metrics = pd.concat([avg_daily_demand, demand_std], axis=1).reset_index()

# Step 5: Calculate Average Lead Time per SKU
# Theory: Lead time is the time it takes for an order to arrive after being placed. It's critical for reorder point calculations.
avg_lead_time = sku_groups["Lead_Time_Days"].mean().rename("Avg_Lead_Time").reset_index()

# Merge lead time data into the metrics DataFrame
metrics = metrics.merge(avg_lead_time, on="SKU_ID", how="left")

# Step 6: Define a function to calculate Safety Stock
# Theory: Safety stock is buffer inventory to protect against uncertainties in demand and lead time.
def calculate_safety_stock(service_level, demand_std_dev, lead_time):
    """
    Calculate safety stock based on the desired service level, demand variability, and lead time.

    Parameters:
    - service_level (float): Desired probability of avoiding stockouts (e.g., 0.95 for 95%).
    - demand_std_dev (float): Standard deviation of daily demand.
    - lead_time (float): Lead time in days.

    Returns:
    - Safety stock (float): Extra inventory units required.
    """
    if demand_std_dev == 0:  # No variability means no safety stock needed.
        return 0
    z = norm.ppf(service_level)  # Convert service level to z-score
    return z * demand_std_dev * np.sqrt(lead_time)  # Safety stock formula

# Apply the safety stock calculation to each SKU
metrics["Safety_Stock"] = metrics.apply(
    lambda row: calculate_safety_stock(desired_service_level, row["Demand_Std_Dev"], row["Avg_Lead_Time"]),
    axis=1
)

# Step 7: Define a function to calculate Reorder Point
# Theory: The reorder point is the inventory level at which a new order should be placed to avoid stockouts.
def calculate_reorder_point(avg_demand, lead_time, safety_stock):
    """
    Calculate reorder point based on average demand, lead time, and safety stock.

    Parameters:
    - avg_demand (float): Average daily demand.
    - lead_time (float): Lead time in days.
    - safety_stock (float): Extra buffer inventory.

    Returns:
    - Reorder point (float): Inventory level at which to reorder.
    """
    return avg_demand * lead_time + safety_stock

# Apply the reorder point calculation to each SKU
metrics["Reorder_Point"] = metrics.apply(
    lambda row: calculate_reorder_point(row["Average_Daily_Demand"], row["Avg_Lead_Time"], row["Safety_Stock"]),
    axis=1
)

# Step 8: Calculate Reorder Quantity using EOQ formula
# Theory: Economic Order Quantity (EOQ) minimizes total inventory costs by balancing holding and ordering costs.
holding_cost_per_unit = 0.5  # Cost to hold one unit of inventory per period
ordering_cost = 20.0  # Fixed cost per order

metrics["Reorder_Quantity"] = np.sqrt(
    (2 * metrics["Average_Daily_Demand"] * ordering_cost) / holding_cost_per_unit
)

# Step 9: Display Results
# Theory: The calculated metrics provide actionable insights for inventory management, such as when to reorder and how much.
print("\nCalculated metrics per SKU:")
print(metrics.head(10))

# **Explanation of New Columns:**
# 1. Average_Daily_Demand: Estimated average daily usage/demand of the SKU.
# 2. Demand_Std_Dev: Variability in demand. Higher variability indicates more unpredictable demand.
# 3. Avg_Lead_Time: Average lead time (days) for the SKU from historical orders.
# 4. Safety_Stock: Extra inventory to protect against uncertainty in demand/lead time.
# 5. Reorder_Point: Inventory level at which a new order should be placed.
# 6. Reorder_Quantity: The recommended order size to balance ordering and holding costs.

# Final Notes:
# - This code is scalable to handle multiple SKUs in a DataFrame.
# - Adjusting the parameters (e.g., desired service level, holding cost, ordering cost) allows for customized inventory strategies.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Simulation parameters
days_to_simulate = 100  # Number of days to simulate
average_demand = 12  # Average daily demand
demand_std_dev = 4  # Standard deviation of daily demand

# Cost parameters
holding_cost_per_unit_per_day = 0.5  # Cost to hold one unit in inventory per day
stockout_cost_per_unit = 5.0  # Cost per unit when stockout occurs
ordering_cost_per_order = 20.0  # Fixed cost per order placed

# Function to generate daily demand using a normal distribution
def generate_demand(days, avg, std):
    # Generate demand and ensure it's at least 1 (no negative or zero demand)
    demand = np.maximum(1, np.random.normal(avg, std, days).astype(int))
    return demand

# Main inventory simulation function
def simulate_inventory(
    days_to_simulate,
    reorder_point,
    reorder_quantity,
    average_demand=12,
    demand_std_dev=4,
    holding_cost_per_unit_per_day=0.5,
    stockout_cost_per_unit=5.0,
    ordering_cost_per_order=20.0,
    initial_inventory=100,
    lead_time_min=2,
    lead_time_max=5
):
    # Generate daily demand
    daily_demand = generate_demand(days_to_simulate, average_demand, demand_std_dev)

    # Initialize inventory, orders, and cost trackers
    inventory = initial_inventory
    orders = []  # Active orders: each is a dict with days_remaining and quantity
    total_cost = 0.0
    total_demand = 0
    total_stockouts = 0

    # Lists to track daily inventory levels and costs
    daily_inventory_levels = []
    daily_costs = []

    for day in range(days_to_simulate):
        demand = daily_demand[day]  # Demand for the current day
        total_demand += demand  # Accumulate total demand

        # Fulfill demand from inventory
        if inventory >= demand:
            inventory -= demand  # Reduce inventory by demand
        else:
            # Stockout occurs if demand exceeds inventory
            stockout_units = demand - inventory
            total_stockouts += stockout_units  # Track stockout units
            inventory = 0  # No inventory left

        # Process incoming orders
        for order in orders[:]:
            order['days_remaining'] -= 1  # Decrease lead time
            if order['days_remaining'] <= 0:  # If order arrives
                inventory += order['quantity']  # Add order quantity to inventory
                orders.remove(order)  # Remove the fulfilled order from the list

        # Place a new order if inventory reaches or falls below reorder point
        if inventory <= reorder_point:
            lead_time = np.random.randint(lead_time_min, lead_time_max + 1)  # Random lead time
            orders.append({'days_remaining': lead_time, 'quantity': reorder_quantity})
            total_cost += ordering_cost_per_order  # Add ordering cost

        # Calculate holding costs for the current day
        holding_cost = inventory * holding_cost_per_unit_per_day
        total_cost += holding_cost

        # Track daily inventory level and total cost
        daily_inventory_levels.append(inventory)
        daily_costs.append(total_cost)

    # Calculate stockout cost
    stockout_cost = total_stockouts * stockout_cost_per_unit
    total_cost += stockout_cost

    # Calculate service level (proportion of demand met without stockouts)
    service_level = 1 - (total_stockouts / total_demand)

    # Return results
    return {
        'service_level': service_level,
        'total_cost': total_cost,
        'daily_inventory_levels': daily_inventory_levels,
        'daily_costs': daily_costs
    }

# Parameters to vary for simulation
reorder_points = [30, 40, 50, 60, 70]  # Different ROL values to test
reorder_quantities = [100, 200, 300]  # Different ROQ values to test

# List to store simulation results
results = []

# Simulate inventory for each combination of ROL and ROQ
for rp in reorder_points:
    for rq in reorder_quantities:
        sim_result = simulate_inventory(
            days_to_simulate,
            reorder_point=rp,
            reorder_quantity=rq,
            average_demand=average_demand,
            demand_std_dev=demand_std_dev,
            holding_cost_per_unit_per_day=holding_cost_per_unit_per_day,
            stockout_cost_per_unit=stockout_cost_per_unit,
            ordering_cost_per_order=ordering_cost_per_order,
        )
        # Store results: ROL, ROQ, service level, total cost
        results.append((rp, rq, sim_result['service_level'], sim_result['total_cost']))

# Convert results to numpy array for easier processing
results_array = np.array(results, dtype=object)

# Plotting the trade-offs between service level and total cost
plt.figure(figsize=(10, 6))
for rp in reorder_points:
    # Filter results for a specific reorder point
    subset = results_array[results_array[:, 0] == rp]
    plt.scatter(subset[:, 2], subset[:, 3], label=f"ROL={rp}")
    for row in subset:
        # Annotate each point with the reorder quantity (ROQ)
        plt.text(row[2], row[3], f"RQ={row[1]}", fontsize=8)

plt.title("Service Level vs Total Cost for Different ROL and ROQ Policies")
plt.xlabel("Service Level")
plt.ylabel("Total Cost")
plt.legend()
plt.grid(True)
plt.show()

# Plotting cost vs service level for each ROL
plt.figure(figsize=(10, 6))
for rp in reorder_points:
    subset = results_array[results_array[:, 0] == rp]
    # Sort by service level for better visualization
    sorted_subset = subset[subset[:, 2].argsort()]
    plt.plot(sorted_subset[:, 2].astype(float), sorted_subset[:, 3].astype(float), marker='o', label=f"ROL={rp}")

plt.title("Cost vs Service Level Across Different ROLs")
plt.xlabel("Service Level")
plt.ylabel("Total Cost")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from typing import Dict, List
from dataclasses import dataclass

# Step 1: Define parameters for inventory management
# Theory: This uses a Python `dataclass` to encapsulate all the constants and inputs required for calculations.
# Computer Science Concept: **Dataclass**
# - A dataclass is a Python feature that simplifies the creation of classes used to store data.
# - It automatically generates an initializer (__init__), making the code cleaner and easier to read.
@dataclass
class InventoryParams:
    annual_demand: float      # D - Annual demand in units
    ordering_cost: float      # S - Cost per order
    holding_cost: float       # H - Annual holding cost per unit
    lead_time_days: int       # L - Lead time in days
    days_per_year: int = 360  # Number of working days per year
    initial_inventory: float = 0.0  # Initial inventory level at the start of the simulation

# Step 2: Define the deterministic inventory model
# Theory: This class models the inventory management process for deterministic demand (demand is known).
# Computer Science Concept: **Class**
# - A class groups related data (attributes) and operations (methods) to model a real-world entity.
# - Here, the `DeterministicInventory` class models an inventory system with attributes like `EOQ` and methods like `simulate`.
class DeterministicInventory:
    def __init__(self, params: InventoryParams):
        """
        Initialize the inventory model with given parameters.

        Parameters:
        - params: An instance of InventoryParams containing constants for simulation.
        """
        self.params = params
        # Calculate Economic Order Quantity (EOQ)
        self.eoq = self._calculate_eoq()  # Computer Science Concept: **Encapsulation** (hidden calculation logic)
        # Daily demand is annual demand divided by working days in a year
        self.daily_demand = params.annual_demand / params.days_per_year  # Simple mathematical operation
        # Reorder point is the inventory level at which a new order should be placed
        self.reorder_point = self.daily_demand * params.lead_time_days  # Derived from lead time and daily demand

        # Initialize variables to track inventory and costs
        self.current_inventory = params.initial_inventory
        # Computer Science Concept: **List of Dictionaries**
        # - A list of dictionaries is used here to store information about pending orders (e.g., quantity, arrival day).
        # - This structure is dynamic and allows for easy updates during the simulation.
        self.pending_orders: List[Dict] = []  # Keeps track of orders in transit
        self.history = {  # Tracks daily metrics for analysis
            'day': [],
            'inventory': [],
            'orders': [],
            'order_costs': [],
            'holding_costs': []
        }

    # Step 3: Define a method to calculate Economic Order Quantity (EOQ)
    # Theory: EOQ minimizes total inventory costs by balancing ordering and holding costs.
    # Computer Science Concept: **Method**
    # - A method is a function that belongs to a class. It operates on the classs attributes and provides behavior.
    def _calculate_eoq(self) -> float:
        """Calculate Economic Order Quantity using Wilson formula"""
        return np.sqrt(
            (2 * self.params.annual_demand * self.params.ordering_cost) /
            self.params.holding_cost
        )  # Computer Science Concept: **Math with Numpy** (efficient numerical computation)

    # Step 4: Define a method to calculate daily costs
    # Theory: Daily costs include ordering costs (if an order is placed) and holding costs for maintaining inventory.
    def _calculate_daily_costs(self, inventory: float, orders: int) -> tuple:
        """Calculate ordering and holding costs for a day"""
        order_cost = orders * self.params.ordering_cost  # Ordering cost = cost per order  number of orders
        holding_cost = (inventory * self.params.holding_cost) / self.params.days_per_year  # Holding cost per day
        return order_cost, holding_cost  # Return both costs as a tuple

    # Step 5: Define the simulation method
    # Theory: This method simulates inventory operations over a given number of days, tracking metrics like inventory levels and costs.
    def simulate(self, num_days: int):
        """Run the simulation for specified number of days"""
        unmet_demand = 0  # Tracks stockouts
        total_demand = 0  # Tracks total demand

        for day in range(num_days):  # Loop through each day in the simulation
            # Computer Science Concept: **Iteration**
            # - Iteration is used here to simulate daily inventory operations over a fixed time horizon.

            # Process any arriving orders
            delivered_quantity = 0
            remaining_orders = []
            for order in self.pending_orders:  # Check pending orders for arrivals
                if order['arrival_day'] == day:
                    delivered_quantity += order['quantity']  # Add delivered quantity to inventory
                else:
                    remaining_orders.append(order)  # Keep other orders pending
            self.pending_orders = remaining_orders

            # Add delivered inventory to current inventory
            self.current_inventory += delivered_quantity

            # Check if we need to place a new order
            orders_placed = 0
            if self.current_inventory <= self.reorder_point:  # Trigger reorder if inventory is below the reorder point
                self.pending_orders.append({
                    'quantity': self.eoq,
                    'arrival_day': day + self.params.lead_time_days
                })
                orders_placed = 1  # Mark that an order has been placed

            # Subtract daily demand
            if self.current_inventory >= self.daily_demand:
                self.current_inventory -= self.daily_demand  # Fulfill daily demand from inventory
            else:
                unmet_demand += self.daily_demand - self.current_inventory  # Record unmet demand
                self.current_inventory = 0  # Inventory goes to zero if demand exceeds it

            total_demand += self.daily_demand

            # Calculate costs for the day
            order_cost, holding_cost = self._calculate_daily_costs(
                self.current_inventory, orders_placed
            )

            # Record daily metrics in the history dictionary
            # Computer Science Concept: **Dictionary Storage** (track multiple metrics for analysis)
            self.history['day'].append(day)
            self.history['inventory'].append(self.current_inventory)
            self.history['orders'].append(orders_placed)
            self.history['order_costs'].append(order_cost)
            self.history['holding_costs'].append(holding_cost)

        # Calculate service level and fill rate
        self.service_level = 1 - (unmet_demand / total_demand)  # Percentage of demand met without stockouts
        self.fill_rate = (total_demand - unmet_demand) / total_demand  # Proportion of demand fulfilled

    # Step 6: Plot service level and fill rate
    # Theory: Visualization helps analyze the simulation results intuitively.
    def plot_service_level(self):
        """Plot service level and fill rate."""
        labels = ['Service Level', 'Fill Rate']
        values = [self.service_level * 100, self.fill_rate * 100]

        plt.figure(figsize=(8, 6))
        plt.bar(labels, values, color=['blue', 'green'], alpha=0.7)
        plt.ylim(0, 100)
        plt.ylabel('Percentage')
        plt.title('Service Level and Fill Rate', fontsize=14)
        plt.grid(axis='y', alpha=0.3)
        for i, v in enumerate(values):
            plt.text(i, v + 2, f'{v:.1f}%', ha='center', fontsize=12)
        plt.show()

    # Step 7: Plot detailed results
    # Theory: Plots include inventory levels, order patterns, and cost breakdowns for a comprehensive analysis.
    def plot_results(self):
        """Create visualizations of the simulation results"""
        df = pd.DataFrame(self.history)  # Computer Science Concept: **DataFrame for structured data storage**

        # Create subplot figure
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle('Deterministic Inventory Simulation Analysis', fontsize=16)

        # Plot inventory level over time
        ax1.plot(df['day'], df['inventory'], 'b-', label='Inventory Level')
        ax1.axhline(y=self.reorder_point, color='r', linestyle='--',
                    label=f'Reorder Point ({self.reorder_point:.0f})')
        ax1.set_title('Inventory Level Over Time')
        ax1.set_xlabel('Day')
        ax1.set_ylabel('Units')
        ax1.grid(True)
        ax1.legend()

        # Plot order pattern
        order_days = df[df['orders'] > 0]['day']
        ax2.vlines(order_days, 0, self.eoq, color='g', label='Orders')
        ax2.axhline(y=self.eoq, color='r', linestyle='--',
                    label=f'EOQ ({self.eoq:.0f})')
        ax2.set_title('Order Pattern')
        ax2.set_xlabel('Day')
        ax2.set_ylabel('Order Quantity')
        ax2.grid(True)
        ax2.legend()

        # Plot cumulative costs
        df['cum_order_costs'] = df['order_costs'].cumsum()
        df['cum_holding_costs'] = df['holding_costs'].cumsum()
        df['total_costs'] = df['cum_order_costs'] + df['cum_holding_costs']

        ax3.plot(df['day'], df['cum_order_costs'], 'b-', label='Ordering Costs')
        ax3.plot(df['day'], df['cum_holding_costs'], 'g-', label='Holding Costs')
        ax3.plot(df['day'], df['total_costs'], 'r-', label='Total Costs')
        ax3.set_title('Cumulative Costs Over Time')
        ax3.set_xlabel('Day')
        ax3.set_ylabel('Cost')
        ax3.grid(True)
        ax3.legend()

        # Plot cost breakdown
        total_order_cost = df['order_costs'].sum()
        total_holding_cost = df['holding_costs'].sum()
        costs = [total_order_cost, total_holding_cost]
        labels = ['Ordering Costs', 'Holding Costs']
        ax4.pie(costs, labels=labels, autopct='%1.1f%%')
        ax4.set_title('Total Cost Breakdown')

        plt.tight_layout()
        plt.show()

        # Print summary statistics
        print("\nSimulation Summary:")
        print(f"EOQ: {self.eoq:.2f} units")
        print(f"Reorder Point: {self.reorder_point:.2f} units")
        print(f"Total Costs: ${df['total_costs'].iloc[-1]:.2f}")
        print(f"Average Inventory: {df['inventory'].mean():.2f} units")
        print(f"Number of Orders: {df['orders'].sum()}")

# Run the simulation
params = InventoryParams(
    annual_demand=10000,    # 10,000 units per year
    ordering_cost=100,      # $100 per order
    holding_cost=20,        # $20 per unit per year
    lead_time_days=5,       # 5 days lead time
    initial_inventory=500   # Start with 500 units
)

simulation = DeterministicInventory(params)
simulation.simulate(num_days=360)  # Simulate for one year
simulation.plot_results()
c
```

```python
import simpy
import random

def warehouse(env, pallets, loading_dock, unload_time, load_time):
    while True:
        # Unload a pallet
        with loading_dock.request() as req:
            yield req
            print(f"Start unloading at {env.now}")
            yield env.timeout(unload_time)
            print(f"End unloading at {env.now}")

        # Process the pallet (e.g., storing in the warehouse)
        yield env.timeout(random.randint(1, 3))

        # Load a pallet for shipping
        with loading_dock.request() as req:
            yield req
            print(f"Start loading at {env.now}")
            yield env.timeout(load_time)
            print(f"End loading at {env.now}")

# Set up the environment
env = simpy.Environment()
loading_dock = simpy.Resource(env, capacity=2)  # Two loading docks
unload_time = 5
load_time = 3

# Add warehouse process
for _ in range(3):  # Three pallets in the system
    env.process(warehouse(env, _, loading_dock, unload_time, load_time))

# Run the simulation
env.run(until=30)
```

```python
import simpy
import random

def customer(env, name, service_desk, service_time):
    """
    A 'customer' process.

    Steps this customer will follow:
    1. Arrive at the service desk.
    2. Request the service desk resource (if busy, must wait).
    3. Once the resource is available, get served for `service_time`.
    4. Leave after being served.

    Parameters:
    - env: The simulation environment
    - name: A string, the customer's name (e.g., "Customer 1")
    - service_desk: The resource (the single server)
    - service_time: The time it takes to serve this customer
    """
    arrival_time = env.now
    print(f"{arrival_time:.2f}: {name} arrives at the desk.")

    # Request the service desk resource
    with service_desk.request() as request:
        yield request  # Wait until the resource is free

        # Now it's this customer's turn to be served
        print(f"{env.now:.2f}: {name} starts being served.")
        yield env.timeout(service_time)  # Service takes some time
        print(f"{env.now:.2f}: {name} leaves after being served.")

def customer_generator(env, service_desk, interarrival_min, interarrival_max, service_min, service_max):
    """
    A process that continuously generates customers.

    Steps:
    1. Wait for a random inter-arrival time before generating the next customer.
    2. Create a new customer process and add it to the environment.
    3. Repeat until simulation ends.

    Parameters:
    - env: The simulation environment
    - service_desk: The resource customers will request
    - interarrival_min, interarrival_max: The range for time between arrivals
    - service_min, service_max: The range for service times
    """
    customer_count = 0
    while True:
        # Wait for a random time before the next customer arrives
        interarrival_time = random.uniform(interarrival_min, interarrival_max)
        yield env.timeout(interarrival_time)

        customer_count += 1
        # Generate a random service time for this new customer
        service_time = random.uniform(service_min, service_max)
        # Start the customer process
        env.process(customer(env, f"Customer {customer_count}", service_desk, service_time))

# ---------------------------
# Main Simulation Setup
# ---------------------------

# Create the simulation environment
env = simpy.Environment()

# Create a service desk resource with a capacity of 1 (only one customer can be served at a time)
service_desk = simpy.Resource(env, capacity=1)

# Parameters for our scenario
INTERARRIVAL_MIN = 5
INTERARRIVAL_MAX = 10
SERVICE_MIN = 4
SERVICE_MAX = 8
SIMULATION_TIME = 50  # Run simulation for 50 minutes

# Start the customer generator process
env.process(customer_generator(env, service_desk, INTERARRIVAL_MIN, INTERARRIVAL_MAX, SERVICE_MIN, SERVICE_MAX))

# Run the simulation
print("Starting the simulation...\n")
env.run(until=SIMULATION_TIME)
print("\nSimulation ended.")
```

```python
import simpy

def inventory_process(env, initial_inventory, daily_demand, reorder_point, reorder_quantity, lead_time):
    """A simple inventory process using SimPy.

    Each day:
    1. Demand consumes some inventory.
    2. If inventory <= reorder point, place an order that arrives after lead_time.
    3. Move to the next day.
    """

    # Initialize inventory
    inventory = initial_inventory

    while True:
        # It's the start of a new 'day'
        print(f"Day {env.now}: Starting inventory = {inventory} units")

        # Satisfy today's demand
        if inventory >= daily_demand:
            # Enough inventory
            inventory -= daily_demand
            print(f"Day {env.now}: Demand of {daily_demand} units met, remaining inventory = {inventory}")
        else:
            # Not enough inventory (some demand goes unmet)
            unmet = daily_demand - inventory
            inventory = 0
            print(f"Day {env.now}: Only partially met demand. Unmet: {unmet} units, inventory now = {inventory}")

        # Check if we need to reorder
        if inventory <= reorder_point:
            print(f"Day {env.now}: Inventory below ROP ({reorder_point}). Placing order for {reorder_quantity} units.")

            # Schedule delivery after lead_time
            def delivery():
                nonlocal inventory
                yield env.timeout(lead_time)
                inventory += reorder_quantity
                print(f"Day {env.now}: Order of {reorder_quantity} units arrived. Inventory now = {inventory}")

            # Start a delivery process
            env.process(delivery())

        # Move on to the next day
        yield env.timeout(1)  # Advance simulation by 1 day

# ---------------------------
# Simulation Parameters
# ---------------------------
initial_inventory = 50
daily_demand = 10
reorder_point = 20
reorder_quantity = 50
lead_time = 2
simulation_days = 10

# Create the environment
env = simpy.Environment()

# Start the inventory process
env.process(inventory_process(env, initial_inventory, daily_demand, reorder_point, reorder_quantity, lead_time))

# Run the simulation
env.run(until=simulation_days)
```

```python
"""

Event-Driven Simulation
Concept: DES operates by simulating events that change the state of a system. Time advances only when an event occurs, making it efficient for modeling systems with intermittent activities.
Implementation in Code:
simpy.Environment: Manages the simulation's timeline and events. Processes like demand_generator and handle_order are event-driven generators that yield env.timeout to model delays.
Events like order arrivals (handle_order) and demand generation (demand_generator) trigger state changes, such as updating inventory levels or placing new orders.
2. State Variables and Metrics
Concept: The state of the system is represented by variables such as inventory levels, costs, and stockouts. These variables are updated as events occur.
Implementation in Code:
InventoryMetrics: Tracks key state variables like inventory_level, costs_level, stockouts, and demand_history.
These metrics are updated in the observe and demand_generator functions after each event, ensuring the simulation accurately reflects the system's state at all times.
3. Concurrency with Generators
Concept: DES models concurrent processes that interact with each other, such as customer demand and order arrivals.
Implementation in Code:
Python Generators: Functions like demand_generator and handle_order use yield to pause and resume execution, simulating real-world concurrency.
Example: handle_order waits for the lead time (yield env.timeout(lead_time)) before updating inventory, while demand_generator continues generating demands.
4. Stochastic Behavior
Concept: Many real-world systems involve randomness, such as demand variability and lead time uncertainty. DES incorporates stochastic elements to capture this variability.
Implementation in Code:
Randomness in Demand and Lead Time:
generate_demand uses np.random.normal to simulate demand based on the mean and standard deviation.
generate_lead_time applies the same approach to model variability in lead times.
These stochastic processes ensure the simulation mimics real-world unpredictability.
5. Performance Metrics and Analysis
Concept: The goal of DES is to evaluate system performance using metrics like service level, average inventory, and total cost.
Implementation in Code:
calculate_statistics:
Aggregates data from InventoryMetrics to compute KPIs such as "Service Level", "Stock Turn Rate", and "Total Cost".
These metrics provide insights into the efficiency and reliability of the inventory system.
Detailed Analysis:
The print_detailed_analysis function summarizes performance metrics and compares them to expected values, helping users assess system behavior.
"""

from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Optional, Union
import numpy as np
import simpy
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

@dataclass
class InventoryParameters:
    """Data class to store inventory system parameters with type hints"""
    reorder_level: int  # Theory: Minimum stock level at which a new order is placed.
    reorder_qty: int  # Theory: Quantity ordered when inventory reaches reorder level.
    purchase_cost: float  # CS Concept: Floating-point type representing the cost per unit of inventory.
    selling_price: float  # Theory: Price at which inventory is sold to customers.
    holding_cost: float  # Theory: Cost of holding one unit of inventory per time unit.
    lead_time_mean: float  # CS Concept: Mean of the normal distribution modeling delivery lead time.
    lead_time_std: float  # CS Concept: Standard deviation of the lead time distribution.
    demand_mean: float  # Theory: Average customer demand per time unit.
    demand_std: float  # Theory: Variability in customer demand, modeled using standard deviation.
    service_level: float  # Theory: Desired service level as a Z-score, representing stockout probability.

@dataclass
class InventoryMetrics:
    """Data class to store running metrics of the simulation"""
    obs_time: List[float] = field(default_factory=list)  # CS Concept: Tracks observation timestamps.
    inventory_level: List[float] = field(default_factory=list)  # CS Concept: Tracks inventory levels over time.
    costs_level: List[float] = field(default_factory=list)  # CS Concept: Tracks cumulative costs over time.
    order_points: List[Tuple[float, float]] = field(default_factory=list)  # Tracks when and at what level orders are placed.
    stockouts: int = 0  # Tracks the number of stockouts during the simulation.
    total_demands: int = 0  # Tracks the total customer demand over time.
    demand_history: List[int] = field(default_factory=list)  # Records demand values at each time step.
    holding_costs: List[float] = field(default_factory=list)  # Tracks the holding costs incurred over time.

class EnhancedInventorySimulation:
    def __init__(self,
                 env: simpy.Environment,
                 params: InventoryParameters) -> None:
        """
        Initialize simulation with parameters and environment.

        CS Concept: **Encapsulation**
        - Encapsulates parameters, inventory levels, metrics, and processes.
        """
        self.params = params  # Theory: Parameters define inventory rules and system behavior.
        self.env = env  # CS Concept: **Simulation Environment** manages simulation time and events.
        self.inventory = params.reorder_qty  # Initial inventory level set to reorder quantity.
        self.num_ordered = 0  # Tracks quantity on order but not yet delivered.
        self.balance = 0.0  # Tracks financial balance (revenue - costs).
        self.metrics = InventoryMetrics()  # Tracks performance metrics during simulation.

        # Theory: Safety stock reduces the risk of stockouts by accounting for demand and lead time variability.
        self.safety_stock = self._calculate_safety_stock()
        # Theory: Optimal reorder point ensures that new orders arrive just in time before stockouts occur.
        self.optimal_rop = self._calculate_reorder_point()

        # CS Concept: **Concurrency**
        # Start generator processes for demand generation and observation.
        self.env.process(self.demand_generator())
        self.env.process(self.observe())

    def _calculate_safety_stock(self) -> float:
        """Calculate safety stock based on lead time and demand variability."""
        # Theory: Safety stock formula accounts for lead time variance and demand variance.
        lt_demand_variance = (
            self.params.lead_time_mean * self.params.demand_std**2 +
            self.params.demand_mean**2 * self.params.lead_time_std**2
        )
        return self.params.service_level * np.sqrt(lt_demand_variance)  # Python Concept: **NumPy** for efficient computation.

    def _calculate_reorder_point(self) -> float:
        """Calculate the optimal reorder point."""
        # Theory: Reorder point = expected demand during lead time + safety stock.
        return (self.params.demand_mean * self.params.lead_time_mean +
                self.safety_stock)

    def generate_demand(self) -> int:
        """Generate demand using a normal distribution."""
        # CS Concept: **Randomness** with NumPy to simulate stochastic demand.
        demand = np.random.normal(
            self.params.demand_mean,
            self.params.demand_std
        )
        return max(0, int(round(demand)))  # Ensures demand is non-negative and integer-valued.

    def generate_lead_time(self) -> float:
        """Generate lead time using a normal distribution."""
        # Theory: Lead time variability is modeled with a normal distribution.
        lead_time = np.random.normal(
            self.params.lead_time_mean,
            self.params.lead_time_std
        )
        return max(0.1, lead_time)  # Truncated to avoid non-positive lead times.

    def handle_order(self):
        """Generator function for handling inventory orders."""
        # Theory: Orders are placed to restore inventory above reorder level.
        self.num_ordered = (self.params.reorder_level + 1 - self.inventory)
        self.num_ordered = np.ceil(
            self.num_ordered / self.params.reorder_qty
        ) * self.params.reorder_qty

        # CS Concept: **Event Logging**
        # Record order placement in simulation metrics.
        self.metrics.order_points.append((self.env.now, self.inventory))

        # Apply costs for the order.
        order_cost = self.params.purchase_cost * self.num_ordered
        self.balance -= order_cost

        # Wait for the lead time before order arrival.
        lead_time = self.generate_lead_time()
        yield self.env.timeout(lead_time)

        # Receive the order and update inventory.
        self.inventory += self.num_ordered
        self.num_ordered = 0

    def observe(self):
        """Generator function to periodically observe and log system state."""
        while True:
            self.metrics.obs_time.append(self.env.now)  # Log current simulation time.
            self.metrics.inventory_level.append(self.inventory)  # Log current inventory level.
            self.metrics.costs_level.append(self.balance)  # Log financial balance.

            # Theory: Holding costs are proportional to inventory level.
            holding_cost = (
                self.inventory *
                self.params.holding_cost *
                0.1
            )
            self.metrics.holding_costs.append(holding_cost)
            self.balance -= holding_cost  # Deduct holding cost from balance.

            yield self.env.timeout(0.1)  # CS Concept: **Time Progression** in simulation.

    def demand_generator(self):
        """Generator function to simulate customer demand."""
        while True:
            # Generate demand and update metrics.
            demand = self.generate_demand()
            self.metrics.total_demands += demand
            self.metrics.demand_history.append(demand)

            # Fulfill demand if inventory is sufficient.
            if demand <= self.inventory:
                revenue = self.params.selling_price * demand
                self.balance += revenue
                self.inventory -= demand
            else:  # Handle stockouts if demand exceeds inventory.
                revenue = self.params.selling_price * self.inventory
                self.balance += revenue
                self.metrics.stockouts += (demand - self.inventory)
                self.inventory = 0

            # Check reorder condition and trigger order if needed.
            if self.inventory <= self.params.reorder_level and self.num_ordered == 0:
                self.env.process(self.handle_order())

            yield self.env.timeout(1.0)  # Wait for the next demand interval.


    def calculate_statistics(self) -> Dict[str, float]:
        """Calculate comprehensive system statistics"""
        inventory_array = np.array(self.metrics.inventory_level)
        demand_array = np.array(self.metrics.demand_history)

        stats = {
            "Average Inventory": float(np.mean(inventory_array)),
            "Inventory Std Dev": float(np.std(inventory_array)),
            "Service Level": 1 - self.metrics.stockouts / self.metrics.total_demands,
            "Average Demand": float(np.mean(demand_array)),
            "Demand Std Dev": float(np.std(demand_array)),
            "Calculated Safety Stock": self.safety_stock,
            "Optimal Reorder Point": self.optimal_rop,
            "Average Lead Time": self.params.lead_time_mean,
            "Lead Time Std Dev": self.params.lead_time_std,
            "Total Cost": -self.balance,
            "Number of Orders": len(self.metrics.order_points),
            "Average Order Size": (
                self.params.reorder_qty if self.metrics.order_points
                else 0
            ),
            "Stock Turn Rate": (
                self.metrics.total_demands /
                float(np.mean(inventory_array)) if len(inventory_array) > 0
                else 0
            )
        }
        return stats

    def print_detailed_analysis(self) -> None:
        """Print comprehensive analysis of the simulation results"""
        stats = self.calculate_statistics()

        print("\nDetailed Inventory System Analysis")
        print("================================")

        print("\n1. Demand Model Parameters:")
        print(f"   - Distribution: Normal(={self.params.demand_mean:.2f}, ={self.params.demand_std:.2f})")
        print(f"   - Observed Average: {stats['Average Demand']:.2f}")
        print(f"   - Observed Std Dev: {stats['Demand Std Dev']:.2f}")

        print("\n2. Lead Time Model:")
        print(f"   - Distribution: Normal(={self.params.lead_time_mean:.2f}, ={self.params.lead_time_std:.2f})")
        print(f"   - Truncated at minimum of 0.1 time units")

        print("\n3. Safety Stock Analysis:")
        print(f"   - Calculated Safety Stock: {stats['Calculated Safety Stock']:.2f}")
        print(f"   - Service Level (Z-score): {self.params.service_level:.2f}")
        print(f"   - Achieved Service Level: {stats['Service Level']:.2%}")

        print("\n4. Reorder Point Analysis:")
        print(f"   - Optimal Reorder Point: {stats['Optimal Reorder Point']:.2f}")
        print(f"   - Current Setting: {self.params.reorder_level}")

        print("\n5. Inventory Performance:")
        print(f"   - Average Inventory: {stats['Average Inventory']:.2f}")
        print(f"   - Inventory Std Dev: {stats['Inventory Std Dev']:.2f}")
        print(f"   - Stock Turn Rate: {stats['Stock Turn Rate']:.2f}")

        print("\n6. Cost Analysis:")
        print(f"   - Total Cost: ${stats['Total Cost']:.2f}")
        print(f"   - Number of Orders: {stats['Number of Orders']}")
        print(f"   - Average Order Size: {stats['Average Order Size']:.2f}")

    def create_dashboard(self):
        """Create and display comprehensive inventory dashboard"""
        dashboard = InventoryDashboard(self)

        # Create all dashboard components
        kpi_cards = dashboard.create_kpi_cards()
        inventory_timeline = dashboard.create_inventory_timeline()
        demand_analysis = dashboard.create_demand_analysis()
        financial_dashboard = dashboard.create_financial_dashboard()

        # Display dashboard
        kpi_cards.show()
        inventory_timeline.show()
        demand_analysis.show()
        financial_dashboard.show()


def run_example():
    # Define parameters
    params = InventoryParameters(
        reorder_level=100,
        reorder_qty=500,
        purchase_cost=50.0,
        selling_price=100.0,
        holding_cost=5.0,
        lead_time_mean=5.0,
        lead_time_std=1.0,
        demand_mean=100.0,
        demand_std=20.0,
        service_level=1.96
    )

    # Create and run simulation
    env = simpy.Environment()
    sim = EnhancedInventorySimulation(env, params)
    env.run(until=100)

    # Print analysis
    sim.print_detailed_analysis()

if __name__ == "__main__":
    run_example()
```

```python
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Optional, Union
import numpy as np
import simpy
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd
from scipy import stats

# [Previous InventoryParameters and InventoryMetrics classes remain the same]

class InventoryDashboard:
    """
    Creates and manages interactive dashboard for inventory simulation results
    """
    def __init__(self, simulation: EnhancedInventorySimulation):
        self.sim = simulation
        self.metrics = simulation.metrics
        self.params = simulation.params

    def create_inventory_timeline(self) -> go.Figure:
        """Create interactive inventory level timeline"""
        df = pd.DataFrame({
            'Time': self.metrics.obs_time,
            'Inventory': self.metrics.inventory_level,
            'Reorder Level': [self.params.reorder_level] * len(self.metrics.obs_time)
        })

        fig = go.Figure()

        # Add inventory level line
        fig.add_trace(go.Scatter(
            x=df['Time'],
            y=df['Inventory'],
            name='Inventory Level',
            line=dict(color='blue', width=2)
        ))

        # Add reorder level line
        fig.add_trace(go.Scatter(
            x=df['Time'],
            y=df['Reorder Level'],
            name='Reorder Point',
            line=dict(color='red', dash='dash')
        ))

        # Add order points
        order_times = [point[0] for point in self.metrics.order_points]
        order_levels = [point[1] for point in self.metrics.order_points]
        fig.add_trace(go.Scatter(
            x=order_times,
            y=order_levels,
            mode='markers',
            name='Order Placed',
            marker=dict(color='green', size=10, symbol='triangle-up')
        ))

        fig.update_layout(
            title='Inventory Level Over Time',
            xaxis_title='Time',
            yaxis_title='Units',
            template='plotly_white',
            hovermode='x unified'
        )

        return fig

    def create_demand_analysis(self) -> go.Figure:
        """Create demand pattern analysis visualization"""
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=(
                'Demand Distribution',
                'Demand Pattern Over Time',
                'Demand vs. Inventory Scatter',
                'Cumulative Demand'
            )
        )

        # Demand distribution
        demand_data = self.metrics.demand_history
        fig.add_trace(
            go.Histogram(
                x=demand_data,
                name='Demand Distribution',
                nbinsx=30,
                showlegend=False
            ),
            row=1, col=1
        )

        # Demand pattern over time
        fig.add_trace(
            go.Scatter(
                x=list(range(len(demand_data))),
                y=demand_data,
                mode='lines',
                name='Demand Pattern',
                showlegend=False
            ),
            row=1, col=2
        )

        # Demand vs. Inventory scatter
        fig.add_trace(
            go.Scatter(
                x=self.metrics.inventory_level[:-1],
                y=demand_data,
                mode='markers',
                name='Demand vs Inventory',
                marker=dict(
                    color=self.metrics.obs_time[:-1],
                    colorscale='Viridis',
                    showscale=True
                ),
                showlegend=False
            ),
            row=2, col=1
        )

        # Cumulative demand
        cumulative_demand = np.cumsum(demand_data)
        fig.add_trace(
            go.Scatter(
                x=list(range(len(cumulative_demand))),
                y=cumulative_demand,
                mode='lines',
                name='Cumulative Demand',
                showlegend=False
            ),
            row=2, col=2
        )

        fig.update_layout(
            height=800,
            title_text='Demand Analysis Dashboard',
            template='plotly_white'
        )

        return fig

    def create_financial_dashboard(self) -> go.Figure:
        """Create financial performance dashboard"""
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=(
                'Cumulative Costs',
                'Cost Breakdown',
                'Order Pattern',
                'Financial KPIs'
            ),
            specs=[[{'type': 'scatter'}, {'type': 'pie'}],
                  [{'type': 'bar'}, {'type': 'indicator'}]]
        )

        # Cumulative costs
        costs_df = pd.DataFrame({
            'Time': self.metrics.obs_time,
            'Total Costs': np.cumsum(self.metrics.holding_costs)
        })

        fig.add_trace(
            go.Scatter(
                x=costs_df['Time'],
                y=costs_df['Total Costs'],
                name='Cumulative Costs',
                showlegend=False
            ),
            row=1, col=1
        )

        # Cost breakdown pie chart
        total_holding_cost = sum(self.metrics.holding_costs)
        total_order_cost = len(self.metrics.order_points) * self.params.purchase_cost

        fig.add_trace(
            go.Pie(
                labels=['Holding Cost', 'Order Cost'],
                values=[total_holding_cost, total_order_cost],
                showlegend=False
            ),
            row=1, col=2
        )

        # Order pattern
        order_counts = pd.Series(
            [point[0] for point in self.metrics.order_points]
        ).value_counts().sort_index()

        fig.add_trace(
            go.Bar(
                x=order_counts.index,
                y=order_counts.values,
                name='Orders',
                showlegend=False
            ),
            row=2, col=1
        )

        # Financial KPI
        service_level = 1 - self.metrics.stockouts / self.metrics.total_demands

        fig.add_trace(
            go.Indicator(
                mode='gauge+number',
                value=service_level * 100,
                title={'text': 'Service Level'},
                gauge={'axis': {'range': [0, 100]}},
                domain={'row': 1, 'column': 1}
            ),
            row=2, col=2
        )

        fig.update_layout(
            height=800,
            title_text='Financial Performance Dashboard',
            template='plotly_white'
        )

        return fig

    def create_kpi_cards(self) -> go.Figure:
        """Create KPI summary cards."""
        stats = self.sim.calculate_statistics()

        # Create figure with subplots for each KPI
        fig = make_subplots(
            rows=1,
            cols=5,  # Fixed number of columns for KPIs
            specs=[[{'type': 'indicator'}] * 5],
            horizontal_spacing=0.1
        )

        kpis = [
            ('Average Inventory', f"{stats['Average Inventory']:.1f}"),
            ('Service Level', f"{stats['Service Level']:.1%}"),
            ('Stock Turn Rate', f"{stats['Stock Turn Rate']:.1f}"),
            ('Total Cost', f"{stats['Total Cost']:,.2f}"),
            ('Number of Orders', f"{stats['Number of Orders']}")
        ]

        # Add each KPI indicator to its own subplot
        for idx, (title, value) in enumerate(kpis):
            try:
                # Clean up the value string
                cleaned_value = value.replace('$', '').replace('%', '').replace(',', '').strip()
                numeric_value = float(cleaned_value)
            except ValueError:
                # Fallback for non-numeric values
                numeric_value = 0.0

            fig.add_trace(
                go.Indicator(
                    mode='number',
                    value=numeric_value,
                    title={'text': title}
                ),
                row=1,
                col=idx + 1  # Columns are 1-based in plotly
            )

        fig.update_layout(
            height=200,
            template='plotly_white',
            showlegend=False
        )

        return fig

def run_example():
    params = InventoryParameters(
        reorder_level=100,
        reorder_qty=500,
        purchase_cost=50.0,
        selling_price=100.0,
        holding_cost=5.0,
        lead_time_mean=5.0,
        lead_time_std=1.0,
        demand_mean=100.0,
        demand_std=20.0,
        service_level=1.96
    )

    env = simpy.Environment()
    sim = EnhancedInventorySimulation(env, params)
    env.run(until=100)

    # Create and display dashboard
    sim.create_dashboard()
    sim.print_detailed_analysis()

if __name__ == "__main__":
    run_example()
```
